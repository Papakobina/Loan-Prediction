```python
# the purpose column has strings. we will use dummy vairables to change them to nunbers since they are categorical
categorical_feats = ['purpose']
final_loans_data =pd.get_dummies(loans_data, columns=categorical_feats, drop_first=True)
```

```python
final_loans_data.head()
```

```python
final_loans_data.columns
```

```python
# train test split
x = final_loans_data[['credit.policy', 'int.rate', 'installment', 'log.annual.inc', 'dti',
       'fico', 'days.with.cr.line', 'revol.bal', 'revol.util',
       'inq.last.6mths', 'delinq.2yrs', 'pub.rec',
       'purpose_credit_card', 'purpose_debt_consolidation',
       'purpose_educational', 'purpose_home_improvement',
       'purpose_major_purchase', 'purpose_small_business']]
y = final_loans_data['not.fully.paid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=101)
```

```python
# implement one decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
single_decision_tree = DecisionTreeClassifier()
```

```python
single_decision_tree.fit(x_train, y_train)
```

```python
# evaluate the model
results_of_single_decision_model = single_decision_tree.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, results_of_single_decision_model))
print(classification_report(y_test, results_of_single_decision_model ))
```

```python
# implement a random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest_model = RandomForestClassifier(n_estimators=200)
```

```python
random_forest_model.fit(x_train, y_train)
```

```python
results_of_random_forest_model = random_forest_model.predict(x_test)
```

```python
print(confusion_matrix(y_test, results_of_random_forest_model))
print(classification_report(y_test, results_of_random_forest_model ))
```

```python
# the random forest outperformed the single decision tree in accuracy by more than 10%
```
